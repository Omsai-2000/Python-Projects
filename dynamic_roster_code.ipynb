```python
import pandas as pd
df=pd.read_excel('Roster_Aug2023.xls')
```

```python
all_row_sliced_data = []

for index, row in df.iterrows():
    if index < 2:
        continue


    row_sliced_data = []

    for column_name, cell_value in row.items():
        if isinstance(cell_value, str):
            sliced_data = cell_value[0:2]
            row_sliced_data.append(sliced_data)


    all_row_sliced_data.append(row_sliced_data)

print(all_row_sliced_data)
```

```python

df2= pd.DataFrame(all_row_sliced_data)
df2
```

```python


def map_shift_code(value):
    if value == 'G1':
        return 'General Shift 1'
    elif value == 'G2':
        return 'General Shift 2'
    elif value == 'D\n':
        return 'Shift D'
    elif value == 'WO':
        return 'Weekly Off'
    else:
        return value

for i in df2[:]:
  df2[i] = df2[i].apply(map_shift_code)
df2
```

```python
from datetime import datetime
df_existing_columns=pd.read_excel('Roster_Aug2023.xls')
columns=df_existing_columns.columns[1:]
data = []
month_mapping = {
    "Jan": "J",
    "Feb": "F",
    "Mar": "M",
    "Apr": "A",
    "May": "M",
    "Jun": "J",
    "Jul": "J",
    "Aug": "A",
    "Sep": "S",
    "Oct": "O",
    "Nov": "N",
    "Dec": "D"
}
for date_str in columns:
    input_date = datetime.strptime(date_str, "%d %b %y %a")
    parts = date_str.split()
    abbreviated_month = parts[1]
    month_prefix = month_mapping.get(abbreviated_month, "Unknown")
    output_date_str = month_prefix + input_date.strftime("%Y-%m-%d")
    data.append({'Date_Column': output_date_str})


df1 = pd.DataFrame(data)
```

```python
df1.reset_index(drop=True, inplace=True)
df3_tran.reset_index(drop=True, inplace=True)

final_df = pd.concat([df1,df3_tran], axis=1)
final_df.columns = range(final_df.shape[1])
final_trans_df=final_df.T
final_trans_df
```
