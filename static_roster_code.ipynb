```python
import pandas as pd
```

```python
from datetime import datetime
```

```python
df_existing_columns=pd.read_excel('Roster_Aug2023.xls')
columns=df_existing_columns.columns[1:32]
data = []
for date_str in columns:
    input_date = datetime.strptime(date_str, "%d %b %y %a")
    output_date_str = "A" + input_date.strftime("%Y-%m-%d")
    data.append({'Date_Column': output_date_str})


df1 = pd.DataFrame(data)
df1
```

```python
row_data = df_existing_columns.loc[2]
sliced_data = row_data.str[0:2]
sliced_data_df = sliced_data.to_frame().reset_index(drop=True)
print(sliced_data_df)
sliced_data_df.drop(0, inplace=True)
df2= pd.DataFrame(sliced_data_df)
df2
```

```python
def map_shift_code(value):
    if value == 'G1':
        return 'General Shift 1'
    elif value == 'G2':
        return 'General Shift 2'
    elif value == 'D\n':
        return 'Shift D'
    elif value == 'WO':
        return 'Weekly Off'
    else:
        return value

df2[2] = df2[2].apply(map_shift_code)
df2 = df2.rename(columns={2: 'Venkat'})
df2
```

```python
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)

merged_df = pd.concat([df1, df2], axis=1)
merged_df.columns = range(merged_df.shape[1])
merged_df
```

```python
data1 = [['EmployeeCode','IN737'],
    ['EmployeeName','Venkat']]

df3 = pd.DataFrame(data1)
df3
transposed_df2 = df3.T
transposed_df2
```

```python
transposed_df = merged_df.T
transposed_df
```

```python
transposed_df2.reset_index(drop=True, inplace=True)
transposed_df.reset_index(drop=True, inplace=True)

final_df = pd.concat([transposed_df2, transposed_df], axis=1)
final_df.columns = range(final_df.shape[1])
final_df
```

```python
# Deleting column if their is any Holidayor Leave

row_indices_to_check = [0, 1]
terms_to_search = ["H", "L"]
mask = (final_df.iloc[row_indices_to_check].isin(terms_to_search)).any()
final_df1 = final_df.loc[:, ~mask]
final_df1
```

```python
output_file = f'RosteringData-venkat.xlsx'
final_df1.to_excel(output_file, index=False,header=False)
```
