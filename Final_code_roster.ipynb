```python
import pandas as pd

df = pd.read_excel('Roster_Aug2023.xls')

all_row_sliced_data = []

for index, row in df.iterrows():
    if index < 2:
        continue

    row_sliced_data = []
    first_column_value = row[df.columns[0]]

    row_sliced_data.append(first_column_value)

    for column_name, cell_value in row.items():
        if isinstance(cell_value, str) and column_name != df.columns[0]:
            sliced_data = cell_value[0:2]
            row_sliced_data.append(sliced_data)

    all_row_sliced_data.append(row_sliced_data)

df2 = pd.DataFrame(all_row_sliced_data)

def map_shift_code(value):
    if value == 'G1':
        return 'General Shift 1'
    elif value == 'G2':
        return 'General Shift 2'
    elif value == 'D\n':
        return 'Shift D'
    elif value == 'WO':
        return 'Weekly Off'
    else:
        return value

for i in df2[:]:
  df2[i] = df2[i].apply(map_shift_code)
df2=df2.sort_values(by=[df2.columns[0]], key=lambda col: col.str[0])
df2=df2.T
df4 = df2.iloc[1:]


from datetime import datetime
df_existing_columns=pd.read_excel('Roster_Aug2023.xls')
columns=df_existing_columns.columns[1:]
data = []
month_mapping = {
    "Jan": "J",
    "Feb": "F",
    "Mar": "M",
    "Apr": "A",
    "May": "M",
    "Jun": "J",
    "Jul": "J",
    "Aug": "A",
    "Sep": "S",
    "Oct": "O",
    "Nov": "N",
    "Dec": "D"
}
for date_str in columns:
    input_date = datetime.strptime(date_str, "%d %b %y %a")
    parts = date_str.split()
    abbreviated_month = parts[1]
    month_prefix = month_mapping.get(abbreviated_month, "Unknown")
    output_date_str = month_prefix + input_date.strftime("%Y-%m-%d")
    data.append(output_date_str)


df3 = pd.DataFrame(data)


df3.reset_index(drop=True, inplace=True)
df4.reset_index(drop=True, inplace=True)

result = pd.concat([df3, df4], axis=1)
result=result.T

result.columns = result.iloc[0]
result = result[1:]


emp_df1 = pd.read_excel('Emp_details.xlsx')
emp_df1 = emp_df1.sort_values(by=[emp_df1.columns[1]], key=lambda col: col.str.split().str[2])
emp_df1 = emp_df1.reset_index(drop=True)

emp_df1.reset_index(drop=True, inplace=True)
result.reset_index(drop=True, inplace=True)
combined_df = pd.concat([emp_df1, result], axis=1)



combinations_by_row = []
for index, row in combined_df.iterrows():

    row_combinations = []


    for col in combined_df.columns:
        row_combinations.append((col, row[col]))


    combinations_by_row.append(row_combinations)

all_modified_combinations = []

for i in range(len(combinations_by_row)):
    row_combinations = combinations_by_row[i]
    df = pd.DataFrame(row_combinations, columns=['Column', 'Value'])
    df = df[~df['Value'].str.contains('H|L')]

    modified_combinations = [(col, val) for col, val in zip(df['Column'], df['Value'])]
    all_modified_combinations.append(modified_combinations)

employee_names = combined_df['EmployeeName'].tolist()
third_names = [name.split()[2] if len(name.split()) >= 3 else '' for name in employee_names]
transposed_dataframes = [pd.DataFrame(data).T for data in all_modified_combinations]

for third_name, df in zip(third_names, transposed_dataframes):
    file_name = f'RosteringData-{third_name}.xlsx'
    df.to_excel(file_name, index=False,header=False)
df
```
